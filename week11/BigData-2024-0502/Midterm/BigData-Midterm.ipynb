```python
'''
題目：1
請造訪 "政府開方資料平台"，以關鍵字 "大專院校各校科系別學生數" 進行查詢。
抓回 112 學年度的資料: https://stats.moe.gov.tw/files/detail/112/112_students.csv。
請以 新北市 內的所有大專院校為範圍:
找出該城市內所有科系名中有 "資訊" 兩個字的科系，將其依 縣市、學校名稱、系名，以及 大學部的男學生總數 與 女學生總數 列出來。
其中，大學部的定義為: 資料中屬於 "D 日"，以及"B 學士"，"B 大學" 或 "B 四技" 才算。
資料請以 HTML 表格的形式呈現出來，並以上述的欄位名稱做為抬頭列(必要時移除不需要的引號)，最後將合成資料存成 "你的學號.html"。
'''
import requests

def show_title(title):
    items = title.split(",")
    for k in range(len(items)):
        print(k, items[k])
        
url = "https://stats.moe.gov.tw/files/detail/112/112_students.csv"
html = requests.get(url)

data = html.text.replace('"', '').split('\n')

show_title(data[0]) # 查看欄位的位置。
city = '臺北市'

header = ['縣市名稱', '學校名稱', '科系名稱', '大學部男生計', '大學部女生計']
         # 26        2          4           8              9
result = [header]  

for d in data[1:]:
    items = d.split(',')
    if len(items) != 28:
        continue
    if city in items[26]: 
        if ('資訊' in items[4]) and ('D 日' in items[5]) and (items[6] in ["B 學士", "B 大學", "B 四技"]):
            result.append([city, items[2], items[4], items[8], items[9]])
            
for r in result:
    print(r)
            

```

```python
'''
題目：2
請將"教育部電子書櫃"第一頁中的電子書圖形抓回來，並將其製成一個HTML檔，呈現相同的電子書圖樣的排列。
請先造訪 教育部電子書櫃 網址: https://stats.moe.gov.tw/bookcase/

做法建議:
1)請先透過 BeautifulSoup 來選取"類別是 datalistImg"的所有 img。
2)接著你需要修正她們的真正網址，也就是在影像的來源前附加上網頁網址，讓我們可以將真正的圖連回我們的網頁。
3)最後將他們放入兩列(row)五行(col)的表格中，並以 "eBook.html" 將合成的內容存檔，合成結果請參考下圖。

你可以在表格之前，使用類似下列的文字，來將樣式加入其中，這樣會讓所有的 img 呈現相同的高度:
  <-- 設定所有圖檔的高度都是 250px。
(* BeautifulSoup 抓出來的內容，必要時，可以透過 str 來轉換成真正的字串。)
'''
import requests
from bs4 import BeautifulSoup
import re

# 教育部電子書櫃
url = "https://stats.moe.gov.tw/bookcase/"

title = "教育部電子書櫃"
print(title)

html = requests.get(url)
# print(html.text)
items = BeautifulSoup(html.text, 'lxml')
imgs = items.select('.datalistImg')

anchors = items.select("a.tooltip[href]")
print(anchors)

# for img in imgs:
#     tostr = f'src="{url}'
# #     print(tostr)
#     img = re.sub('src="', tostr, str(img))
#     print(img)

tostr = f'src="{url}'
for k in range(len(imgs)):
    imgs[k] = re.sub('src="', tostr, str(imgs[k]))
    
# 填入表格中:
html = "<style> img { height:250px; } </style> "+ \
       "<table border=1>"
index = 0
for k in range(len(imgs)):
    if index % 5 == 0:
        html += "\n<tr>"
    html += f"\n <th>{str(imgs[k])}</th>"
    index += 1
    if index % 5 == 0:
        html += "\n</tr>"
        
if index % 5 != 0:
    html += "\n</tr>"
else:
    html += "\n</table>"
    
print(html)
```

```python
'''
題目：3
承上題，請先透過 BeautifulSoup 來選取 a 且指定類別是 tooltip (a.tooltip) 的所有 anchors。
接著請模仿上述的修改動作，讓連結同樣分成兩列五行，同時在點選電子書對應的影像圖形後，可以連到電子書所在的網頁。
'''
import requests
from bs4 import BeautifulSoup
import re

# 教育部電子書櫃
url = "https://stats.moe.gov.tw/bookcase/"

title = "教育部電子書櫃"

html = requests.get(url)
# print(html.text)
items = BeautifulSoup(html.text, 'lxml')

# imgs = items.select(".datalistImg")
anchors = items.select("a.tooltip")

for k in range(len(anchors)):
    anchor = str(anchors[k])
    anchor = re.sub('href="', f'href="{url}', anchor)
    anchors[k] = re.sub('src="', f'src="{url}', anchor)

    
html = "<style> img { height:250px; } </style> "+ \
       "<table border=1>"
index = 0
for k in range(len(anchors)):
    if index % 5 == 0:
        html += "\n<tr>"
    html += f"\n <th>{str(anchors[k])}</th>"
    index += 1
    if index % 5 == 0:
        html += "\n</tr>"
        
if index % 5 != 0:
    html += "\n</tr>"
else:
    html += "\n</table>"
    
print(html)
```
