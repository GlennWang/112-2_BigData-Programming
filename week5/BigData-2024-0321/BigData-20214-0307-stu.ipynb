```python
'''
sorted(data, key=picking_rule, inverse=Flase): 
將 data 進行排序，而排序則是依據 key 欄位指定的 picking_rule，回傳排序的結果。
預設是 由小到大 地進行排序，如果要改成 由大到小，則令 inverse=True。
'''
# sorted(.):
import random as rand

def show_by_row(data):
    for d in data:
        print(d)

namestr = "李彥智、張逸凡、劉建岳、李國偉、張淑雯"
names = namestr.split('、')

scores = []
for i in range(len(names)):
    s = [ names[i] ]  # 先設定姓名。
    s += [rand.randint(0, 99) for k in range(3)]  # 後追加三科的成績。
    scores.append(s)
    
print("隨機生成的成績資料:")
show_by_row(scores)
print()

# 針對 第一欄 的成績來排序:
print("針對 第一欄 的成績來排序:")
result_col_1 = sorted(scores, key=lambda x: x[1])  # 取出每筆資料中的第二個元素來進行比較排序。
show_by_row(result_col_1)
print()

# field_1 = [[x[0], x[1]] for x in scores]
field_1 = eval('[[x[0], x[1]] for x in scores]')  # OK!
show_by_row(field_1)
print()

field_1_index = [field_1[k]+[k] for k in range(len(field_1))]  # append()
show_by_row(field_1_index)
print()

result = sorted(field_1_index, key=lambda x: x[1])  # 取出每筆資料中的第二個元素來進行比較排序。
show_by_row(result)
index = [x[2] for x in result]
print(index)
```

```python
'''
動態給定生成的內容。
以f-string輸出的形式中，需要引用變數或運算式的值，可以大括號將其括住。
如果大括號內還需要再引用其他變數時，就再為其加上一層大括號。
'''
# 動態給定生成內容-1
# 將輸出格式動態地加入格式表示式中:
def triangle_R(n):
    fmt = f">{n}"
    pattern = ""
    for k in range(n, 0, -1):
        pattern += str(k)
        print(f"{pattern:{fmt}}")  # 動態設定輸出的格式與引用介紹。
        
triangle_R(5)
```

```python
# 動態給定生成內容-2
import random as rand

# 讓資料逐列顯示方便觀察的函式:
def show_by_row(data):
    for d in data:
        print(d)

# 動態合成取出二維串列中的部分欄位的程式寫法:
# 類似如右的輸出: [ [x[0], x[2]] for x in data ]
def get_columns(data, fields):  # fields 的順序會決定取出的結果。
    fld = "["
    for f in fields:
        if fld == "[":
            fld += f"x[{f}]"
        else:
            fld += f", x[{f}]"
    fld += ']'
    cmd = f"[{fld} for x in data]"
#     print(fld)
    return eval(cmd)
    
        
# 拆解人名資料
namestr = "李彥智、張逸凡、劉建岳、李國偉、張淑雯"
names = namestr.split('、')

# 人名與三科成績的合成
scores = []
for i in range(len(names)):
    s = [ names[i] ]  # 先設定姓名。
    s += [rand.randint(0,99) for k in range(3)]  # 後追加三科的成績。
    scores.append(s)
    
# 顯示生成的成績資料:    
show_by_row(scores)
print()

# 選擇要取出的部分欄位: 以 0 為起始的索引串列。
fields = [0, 2, 3]  # 希望依序取出的欄位。
part = get_columns(scores, fields)
show_by_row(part)
print()
```

```python
'''
將訊息進行加密: 替代式密碼。
'''
# 生成所有的英文字，不分大小寫 => 一律改成小寫，來進行後續的判斷。
import random as rand
# 首先生成轉換表:
chars = [chr(ord('a') + ch) for ch in range(26)]

rand.shuffle(chars)
table = chars

code = ord('a')
result = ""
for c in table:
    if result == "":
        result = f"{chr(code)}->{c}"
    else:
        result += f"\n{chr(code)}->{c}"
    code += 1
        
print(result)
```
