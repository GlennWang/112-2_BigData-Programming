```python
'''
工具函式:
'''
def newline(row=1):
    print('\n' * row, end='')
    
def show(strcmd):
    print(strcmd, '=', eval(strcmd))
```

```python
'''
list 的方法:
append:
clear:
copy:
count:
extend:
index:
insert:
pop:
remove:
reverse:
sort:
'''
    
lst = [1, 2, 3]
show('lst')
newline()

lst.append(5)
print("lst.append(5)")
show('lst')
newline()

lst.append([6, 7])  # 括號中的內容，會被當成是一個元素附加進來
print("lst.append([6, 7])")
show('lst')
newline()

value = lst.pop()
print("value = lst.pop()")
show('lst')
show('value')
newline()

lst.extend([8, 9])  # for list parameter only, 可以一次加入很多項！
print("lst.extend([8, 9])")
show('lst')
newline()

lst.reverse()  # 會將串列(lst)的內容左右翻轉！
print("lst.reverse()")
show('lst')
newline()

cnt = lst.count(3)
print("cnt = lst.count(3)")
show('cnt')
show('lst')
newline()

index = lst.index(5)
print("index = lst.index(5)")
show('index')
newline()

lst.insert(1, 99)
print("lst.insert(1, 99)")
show('lst')
newline()

del lst[0]
print("del lst[0]")
show('lst')
newline()

del lst[-2:]
print("del lst[-2:]")
show('lst')
```

```python
'''
lambda function:
lambda x: x*x
可以用來生成簡短的函式，不需要透過 def 來定義。
'''

double = lambda x : 2 * x
print( double(3) )

mul = lambda x, y : x * y
print( mul(3, 7) )
```

```python
'''
map(fun_name, list_content)
'''

lst = [1, 2, 3, 4, 5]
result = map(str, lst)  # 會回傳將後方串列lst的內容，逐項代入str後的結果。

print(result)  # 目前還是 map object 的形式！
print( list(result) )  # 轉換成串列，才會看到其結果。
print()

# 搭配使用 lambda 函式:
result = map(lambda x: x*x, lst)
print( list(result) )  # 轉換成串列，才會看到其結果。
print()

result = map(lambda x: 2*x, lst)
print( list(result) )
```

```python
'''
請使用Python撰寫遞迴版的 fibonacci number 計算函式: fib(n)。
並估算求得 fib(36) 所需的計算時間。

同上，請示用串列的形式，透過定義式以逐項加入的方式，計算到 fib(36)，並顯示該串列的結果。
'''
import time

def fib(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

n = 1
show('n')

show('fib(n)')
```

```python
'''
numstr = "18, 32, 45, 68, 98, 46, 92, 77, 89, 55, 42, 60, 52, 41, 96, 32, 84, 26, 83, 36"
請先對上述字串進行拆解，接著請轉換成數字串列，並印出該數字串列。
同上，接著請建立一個空串列，再利用insert與append，將數值小於 60 的數字放在前面，大於或等於 60 數字的放在後面。
只要進行數值分類，不必對其內容進行排序。
'''
```

```python
import time
numstr = "18, 32, 45, 68, 98, 46, 92, 77, 89, 55, 42, 60, 52, 41, 96, 32, 84, 26, 83, 36"

times = 100000
start = time.time()
for r in range(times):
    nums = []
    yes = []
    no = []
    nums = map(int, numstr.split(', '))
    for it in nums:
        if it > 70:
            yes.append(it)
        else:
            no.append(it)
stop = time.time()
print("dt=", stop-start)
print(nums)
print(yes)
print(no)
```

```python
import time
numstr = "18, 32, 45, 68, 98, 46, 92, 77, 89, 55, 42, 60, 52, 41, 96, 32, 84, 26, 83, 36"

times = 100000
start = time.time()
for r in range(times):
    nums = [int(it) for it in numstr.split(', ')]
    yes  = [y for y in nums if y > 70]         
    no   = [n for n in nums if n <= 70]   

stop = time.time()
print("dt=", stop-start)
print(nums)
print(yes)
print(no)
```
